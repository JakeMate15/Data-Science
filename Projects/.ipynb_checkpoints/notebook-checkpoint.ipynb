## Creating a DataFrame from the data

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Get the csv file info
netflix_df = pd.read_csv('netflix_data.csv', index_col = 0)
print(netflix_df[:5])
```

```python


netflix_subset = netflix_df[netflix_df.type == 'Movie']

netflix_movies = netflix_subset.loc[: , ['title', 'country', 'genre', 'release_year', 'duration']]

short_movies = netflix_movies[netflix_movies.duration < 60]

colors = []
for lab, movie in netflix_movies.iterrows():
    genre = movie['genre']
    if genre == 'Children':
        colors.append('blue')
    elif genre == 'Documentaries':
        colors.append('red')
    elif genre == 'Stand-Up':
        colors.append('green')
    else :
        colors.append('black')

fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies.loc[:, 'release_year'], netflix_movies.loc[:, 'duration'], c = colors, alpha = .9)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()

answer = 'maybe'
```

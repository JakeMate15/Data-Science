# MatplotLib

Matplotlib es una biblioteca de visualización de datos en Python que proporciona una variedad de gráficos como líneas, barras, dispersión, entre otros. Es ampliamente utilizado debido a su simplicidad y flexibilidad.

```python
import matplotlib.pyplot as plt
```

## Gráfico de Líneas

Un gráfico de líneas se utiliza generalmente para visualizar tendencias a lo largo del tiempo.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

# Crear gráfico de líneas
plt.plot(x, y)
plt.title("Gráfico de Líneas Básico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## Gráfico de Barras
Ideal para comparaciones entre categorías.


```python
# Datos de ejemplo
categorias = ['Categoría A', 'Categoría B', 'Categoría C']
valores = [5, 7, 3]

# Crear gráfico de barras
plt.bar(categorias, valores)
plt.title("Gráfico de Barras Básico")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

## Gráfico de Dispersión

Utilizado para observar relaciones entre dos variables.

```python
# Datos de ejemplo
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

# Crear gráfico de dispersión
plt.scatter(x, y)
plt.title("Gráfico de Dispersión Básico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

# Histograma

Los histogramas son útiles para visualizar la distribución de una variable numérica.

```python
# Datos de ejemplo
datos = [1, 2, 2, 3, 3, 3, 3, 4, 4, 5]

# Crear histograma
plt.hist(datos, bins=5, edgecolor='black')
plt.title("Histograma Básico")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

# Personalización
Matplotlib permite una amplia personalización, incluyendo colores, tipos de líneas, y más.

```python
# Gráfico de líneas con personalización
plt.plot(x, y, color='red', linestyle='--')
plt.title("Gráfico de Líneas Personalizado")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.show()
```

```python
# Datos de ejemplo
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]
sizes = [210, 410, 312, 214, 415, 312, 213, 210, 410, 312, 214, 415, 312]

# Crear gráfico de dispersión personalizado
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title("Gráfico de Dispersión Personalizado")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

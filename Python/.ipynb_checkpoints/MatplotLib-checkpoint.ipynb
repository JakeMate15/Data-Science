# MatplotLib



````python
import matplotlib.pyplot as plt
````

## Basic plots

The plot funciton draws a line between each point. <br>

```
plt.plot(x, y)
```

On the other hand scatter doesn't connect the points

```
plt.scatter(x, y)
```

````python
year = [1950, 1970, 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, population)
plt.show()

plt.scatter(year, population)
plt.show()
````

## Histogram



````python
rep = [5, 5, 5, 5, 12, 10, 23, 12 ,3, 12, 0, 4]

plt.hist(rep, bins = 8)
plt.show()
````

## Customization

* For axis labels
```
plt.xlabe('')
plt.ylabel('')
```

* For titles 
```
plt.title('')
```

* For ticks <br>
    Where the first list are the ticks you want and the second are their names
```
plt.ticks([], [])
```

````python
year = [1950, 1970, 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]

plt.title('World Population')
plt.xlabel('Year')
plt.ylabel('Population')

plt.yticks([1, 3, 5, 7, 9], ['0', '3B', '5B', '7B', '9B'])

plt.plot(year, population)
plt.show()
````
